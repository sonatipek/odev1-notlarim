```python
# Yükleme (Gerekirse)
#? pip install seaborn

# Gerekli Kütüphaneleri İçe Aktarma
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Seaborn'un içindeki örnek veri kümelerini listeleme
sns.get_dataset_names()

# Örneğin 'tips' veri setini yükleyelim
df = sns.load_dataset("tips")
df.head()
```

```python
#  Dağılım Grafiği (scatterplot)
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

```python
# Histogram (histplot)
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.show()
```

```python
# Kutu Grafiği (boxplot)
sns.boxplot(x="day", y="total_bill", data=df)
plt.show()
```

```python
# Bar Grafiği (countplot)
sns.countplot(x="day", data=df)
plt.show()
```

```python
# Çubuk Grafiği (barplot)
sns.barplot(x="day", y="total_bill", data=df, estimator=sum)
plt.show()
```

```python

#Isı Haritası (heatmap)
corr = df.corr(numeric_only=True) 
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```
